```python
import os, time, json, sys
train_data = []
for file in os.listdir(r"P:\1IEC\AISamSung2023\Chấm Bài\TestData_nolabel"):
   with open(os.path.join(r"P:\1IEC\AISamSung2023\Chấm Bài\TestData_nolabel", file), "r") as f:
      file_content = f.read()
   content = {
      "text" : file_content.strip(),
      "label": None,
      "file_name": file
   }
   train_data.append(content)
for file in os.listdir(r"P:\1IEC\AISamSung2023\Chấm Bài\TestData_withlabel\notspamTest"):
   with open(os.path.join(r"P:\1IEC\AISamSung2023\Chấm Bài\TestData_withlabel\notspamTest", file), "r") as f:
      file_content = f.read()
   for i in range(len(train_data)):
      if train_data[i]['text'] == file_content.strip():
         train_data[i]['label'] = "notspam"
         break
for i in range(len(train_data)):
   if train_data[i]['label'] is None:
      train_data[i]['label'] = "spam"
```
