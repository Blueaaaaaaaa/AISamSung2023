```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

train_data = pd.read_csv('Train_samsung.csv')
test_data = pd.read_csv('Test_samsung_noclass.csv')
```

```python
print(train_data.head())
print(train_data.info())
print(train_data.describe())
```

```python
train_data = train_data.dropna()
test_data = test_data.fillna(method='ffill')
```

```python
# Chuyển đổi categorical data nếu có
label_encoders = {}
for column in train_data.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    train_data[column] = label_encoders[column].fit_transform(train_data[column])

# Áp dụng cùng LabelEncoder cho dữ liệu test
for column in test_data.select_dtypes(include=['object']).columns:
    if column in label_encoders:
        test_data[column] = label_encoders[column].transform(test_data[column])
    else:
        # Tạo mới LabelEncoder cho cột không có trong train
        le = LabelEncoder()
        test_data[column] = le.fit_transform(test_data[column])
```

```python
# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_train = scaler.fit_transform(train_data.drop('Class', axis=1))  # Thay 'target_column' bằng cột target thực tế
y_train = train_data['Class']  # Thay 'target_column' bằng cột target thực tế
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
X_test = scaler.transform(test_data)
```

```python
predictions = model.predict(X_test)
```

```python
result = pd.DataFrame({'prediction': predictions})
result.to_csv('predictions.csv', index=False)
print("Dự đoán đã được lưu vào file predictions.csv")
```
